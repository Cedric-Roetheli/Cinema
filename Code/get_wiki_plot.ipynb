```python
import wikipedia
import pandas as pd
from pathlib import Path

# === Load your final matches CSV ===
final_matches = pd.read_csv("final_matches_updated2.csv")

# === Create folder to save summaries ===
summary_output_dir = Path.home() / "Desktop" / "processed_subs" / "summaries"
summary_output_dir.mkdir(parents=True, exist_ok=True)

# === Function to fetch plot summary with smart title fallbacks ===
def fetch_wikipedia_summary(title, year=None):
    attempts = [title]

    if year:
        attempts.append(f"{title} (film)")
        attempts.append(f"{title} ({year} film)")
    else:
        attempts.append(f"{title} (film)")

    for attempt in attempts:
        try:
            page = wikipedia.page(attempt, auto_suggest=False)
            content = page.content

            if "== Plot ==" in content:
                start = content.find("== Plot ==") + len("== Plot ==")
                end = content.find("==", start)
                plot = content[start:end].strip() if end != -1 else content[start:].strip()
                return plot
            else:
                return content.split('\n')[0]

        except Exception as e:
            continue

    print(f"âŒ No summary found for: {title}")
    return None

# === Loop through final matches and save summaries ===
# Change .head(10) to .iterrows() to run on full dataset
for idx, row in final_matches.iterrows():
    title = row["matched_title"]
    filename = row["subtitle_filename"]

    # Skip rows with missing or invalid filenames
    if not isinstance(filename, str) or not filename[:4].isdigit():
        print(f"âš ï¸ Skipping row with bad or missing filename: {filename}")
        continue

    year = filename[:4]

    # Skip if summary already exists
    summary_path = summary_output_dir / f"{filename}_summary.txt"
    if summary_path.exists():
        print(f"ğŸŸ¡ Summary already exists for: {title}")
        continue

    summary = fetch_wikipedia_summary(title, year)
    if summary:
        with open(summary_path, "w", encoding="utf-8") as f:
            f.write(summary)
        print(f"âœ… Saved summary for: {title}")

```
